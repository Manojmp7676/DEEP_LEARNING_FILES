```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
```

```python
x=np.array([[0,0],[0,1],[1,0],[1,1]])#input data for XOR problem
y=np.array([[0],[1],[1],[0]])#output data for XOR problem
```

```python
model=Sequential([
Input(shape=(2,)), #input layer
Dense(4,activation="sigmoid"), #hidden layer
Dense(1,activation="sigmoid")
 ]) #output layer
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])#compiling the model
```

```python
model.fit(x, y, epochs=5000, verbose=0)#training the model
```

```python
pred=model.predict(x)#making predictions
print("Predictions:")#printing the predictions
print(pred)#printing the predicted classes
```
